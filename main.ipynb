```python
from sklearn.datasets import load_breast_cancer
import numpy as np 


df = load_breast_cancer()
x = df.data
y = df.target
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x
```

Test and Train

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3)

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Calculator metrics def

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def caluculator(y_train , y_test, y_pred_train, y_pred_test ): 

    acc_train = accuracy_score(y_true=y_train, y_pred=y_pred_train)
    acc_test = accuracy_score(y_true=y_test, y_pred=y_pred_test)

    precision = precision_score(y_true=y_test, y_pred=y_pred_test)

    recall = recall_score(y_true=y_test, y_pred=y_pred_test)

    print( 'acc_train = ' , acc_train)
    print( 'test = ' , acc_test)
    print( 'precision = ' , precision)
    print( 'recall = ' , recall)

    return acc_train, acc_test, precision, recall
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train , Y_train)
```

```python
y_pred_train = gnb.predict(X_train)
y_pred_test = gnb.predict(X_test)
```

```python
acc_train_gnb, acc_test_gnb, precision_gnb, recall_gnb = caluculator(Y_train, Y_test, y_pred_train, y_pred_test)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
```

```python
knn_pred_train = knn.predict(X_train)
knn_pred_test = knn.predict(X_test)
```

```python
acc_train_knn, acc_test_knn, precision_knn, recall_knn = caluculator(Y_train, Y_test, knn_pred_train, knn_pred_test)
```

Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier

dt  = DecisionTreeClassifier(max_depth=30 , min_samples_split=5)
dt.fit(X_train, Y_train)
```

```python
dt_pred_train = dt.predict(X_train)
dt_pred_test = dt.predict(X_test)
```

```python
acc_train_dt , acc_test_dt, precision_dt, recall_dt = caluculator(Y_train, Y_test, dt_pred_train, dt_pred_test)
```
